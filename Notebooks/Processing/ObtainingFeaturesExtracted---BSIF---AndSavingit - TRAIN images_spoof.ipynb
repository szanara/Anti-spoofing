```python
files = os.listdir(path)
```

```python
tokens = [i for i in range(len(files))]
```

```python
for fil,token in zip(files,tokens):
    read = ReadImage(path, fil)
    image = read.read_file()
    process = ProcessingImage(image)
    features=[]
    if process.extractor() is None:
        continue
    else:
        [ch1_hsv, ch2_hsv, ch3_hsv],[ch1_ycbcr, ch2_ycbcr, ch3_ycbcr] =process.extractor()
        for channel in [ch1_hsv, ch2_hsv, ch3_hsv,ch1_ycbcr, ch2_ycbcr, ch3_ycbcr]:
            Bsif  = BSIF(channel)
            bsif_features = Bsif.bsif()
            features.append(bsif_features)
        features = np.concatenate(features,axis=2)
        features = np.concatenate(features, axis=1)
        print(features.shape)
        np.save(path2save+f'/features_all_bsif_{token}.npy', features)
```
